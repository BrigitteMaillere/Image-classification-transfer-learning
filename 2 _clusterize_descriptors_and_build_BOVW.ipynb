```python
import pandas as pd
import numpy as np
from sklearn.cluster import MiniBatchKMeans
```

```python
random_state = 0
nb_breeds = 10
n_clusters = 1000
results_path = '../'
```

```python
X_train = pd.read_csv(results_path + 'df_train_' + str(nb_breeds) + '_descriptor.csv')
X_test = pd.read_csv(results_path + 'df_test_' + str(nb_breeds) + '_descriptor.csv')

df_train = pd.read_csv(results_path + 'df_train_' + str(nb_breeds) + '_breed_filename.csv')
df_test = pd.read_csv(results_path + 'df_test_' + str(nb_breeds) + '_breed_filename.csv')

y_train = np.load(results_path + 'y_train_' + str(nb_breeds) + '_breeds.npy')
y_test = np.load(results_path + 'y_test_' + str(nb_breeds) + '_breeds.npy')

X = X_train.drop(columns = ['filename']).sample(frac=0.1, random_state = random_state)

```

```python
results_path
```

```python
df_train = pd.read_csv(results_path + 'df_train_' + str(nb_breeds) + '_breed_filename.csv')
df_test = pd.read_csv(results_path + 'df_test_' + str(nb_breeds) + '_breed_filename.csv')
```

```python
df_train.head()
```

```python
km = MiniBatchKMeans(n_clusters=n_clusters, random_state=random_state)
km.fit(X)
```

```python
X_train['visual_word'] = km.predict(X_train.drop(columns=['filename']))
X_test['visual_word'] = km.predict(X_test.drop(columns=['filename']))
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse=False, handle_unknown="ignore")
X_train_encoded = ohe.fit_transform(X_train[['visual_word']])
X_test_encoded = ohe.transform(X_test[['visual_word']])

X_train_encoded = pd.DataFrame(X_train_encoded)
X_test_encoded = pd.DataFrame(X_test_encoded)



```

```python
X_train_encoded['filename'] = X_train.filename.values
X_test_encoded['filename'] = X_test.filename.values

BOVW_train = X_train_encoded.groupby(['filename'], as_index=False).sum()
BOVW_test = X_test_encoded.groupby(['filename'], as_index=False).sum()
```
